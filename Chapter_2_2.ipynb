Imagine the stick lies on the $x$ axis of the standard coordinate system, then, to break the stick in $3$ pieces we only need to choose $2$ break points, let the coordinates of these break points be $A, B$. 

Clearly $0 \leq A,B \leq 1$. Also it has been mentioned in the problem that the break points $A, B$ are chosen at random, which means, $A,B \sim Uniform([0,1])$

The lengths of the sticks after breaking will be 
1. $A$
2. $B-A$
3. $1-B$

Let $E$ be the event that the resulting sticks must form a triangle, for this to happen, the sum of the lengths of any two sides must be greater than the length of the remaining side, more formally,

$$E = A + B - A < 1- B \text{ and } B-A + 1-B < A \text{ and } 1-B+A < B-A$$

$$E = B < \frac{1}{2} \cap A > \frac{1}{2} \cap B > A + \frac{1}{2}$$



$$P(E) = \frac{Area(E)}{Area(S)}$$

$S$ denotes the sample space, the length of the triangle should be positive, therefore the whole unit square will not be the sample space, instead the sample space will be as below
$$S = A > 0 \cup B-A > 0 \cup 1-B > 0$$

$$P(E) = \frac{\frac{1}{8}}{\frac{1}{2}}=\frac{1}{4}$$

```
lim = 1.5


plt.figure(figsize=(5,5))
for p1, p2 in [[[0, 0], [0, 1]], [[0, 1], [1, 1]], [[1, 1], [1, 0]], [[1, 0], [0, 0]]]:
    plt.plot(p1, p2, color='black')

plt.plot([-lim, lim], [-lim+0.5, lim+0.5], ls='--', color='black')

plt.plot([-lim, lim], [-lim, lim], ls='--', color='black')
plt.fill_between(np.linspace(0, 1, 100), np.linspace(0, 1, 100), 
                 1, facecolor='blue', interpolate=True, label='S')
plt.fill_between(np.linspace(0, 0.5, 100), 0.5, 
                 0.5+ np.linspace(0,0.5,100), facecolor='red', interpolate=True, label='E')

plt.axvline(0.5, ls='--', color='black')
plt.axhline(0.5, ls='--', color='black')

plt.annotate("B = 0.5", xy=(1.5, 0.5), xytext=(1.5, 0),
            arrowprops=dict(arrowstyle="->"))
plt.annotate("A = 0.5", xy=(0.5, -0.5), xytext=(-1.5, -1.5),
            arrowprops=dict(arrowstyle="->"))
plt.annotate("B = A + 0.5", xy=(-1, -0.5), xytext=(-0.5, -0.5),
            arrowprops=dict(arrowstyle="->"))
plt.annotate("B = A", xy=(0, 0), xytext=(-1, 1),
            arrowprops=dict(arrowstyle="->"))

plt.xlim([-lim, lim])
plt.ylim([-lim, lim])
plt.legend()
plt.grid()
plt.show()
```

Let the stick lie on the standard coordinate system on $x-$axis and let $x$ be the first break point, then there are two possibilities
1. $x <0.5$
2. $x \geq 0.5$

The conditions will correspond to the plots below

```
fig, axes = plt.subplots(1,2, figsize=(10,5))
x1,y1 = [0.3,0.7]
x2,y2 = [0.8,0.2]
points = [[x1,y1], [x2,y2]]
for idx, ax in enumerate(axes):
    ax.plot([0,1], [0,0], ls='-', color='black', linewidth=4)
    ax.scatter(points[idx], [0,0], marker='X', s=100, color='red')
    ax.annotate('x', xy=(points[idx][0],0), xytext=(points[idx][0],-0.02),
                arrowprops=dict(arrowstyle="->"))
    ax.annotate('y', xy=(points[idx][1],0), xytext=(points[idx][1],-0.02),
                arrowprops=dict(arrowstyle="->"))
    ax.grid()
    ax.set_title('CASE {}'.format(idx+1))
    ax.axvline(0.5, ls='--')
plt.show()
```

If the coordinate system in case 2 are updated such that origin is at 1 and points to the left are positive, then case 2 is same as case 1, therefore we only need to solve for one of the cases

**CASE 1**

Let $X$ denote the first break point and $Y$ denote the second break point, then according to the problem

$$X \sim U(0, 0.5)$$
$$Y|X=x \sim U(x, 1)$$

$$f_X(x) = \begin{cases} 1 \quad; 0 \leq x \leq 0.5 \\ 0 \quad; \text{else} \end{cases}$$

$$f_{Y|X=x}(y) = \begin{cases} \frac{1}{1-x} \quad; x \leq y \leq 1 \\ 0 \quad \quad ; \text{else} \end{cases}$$


$$f_{XY}(x,y)=f_X(x)f_{Y|X=x}(y) = \begin{cases} \frac{1}{1-x} \quad; x \leq y \leq 1, 0\leq x\leq 0.5 \\ 0 \quad \quad; \text{else}\end{cases}$$

Let $E$ be the event that the pieces form a traingle, then $E$ wil be given the region
$$x \leq 0.5, 0.5 \leq y \leq 0.5+x$$

$P(E)=\frac{f_{XY}(E)}{f_{XY}(\Omega)}$, where $\Omega$ is the sample space

$$f_{XY}(E) = \int_{0}^{0.5} \int_{0.5}^{0.5+x}\frac{1}{1-x}dy dx = \int_{0}^{0.5} \frac{x}{1-x}dx = \frac{2log2-1}{2}$$

$$f_{XY}(\Omega) = \int_{0}^{0.5} \int_{x}^{1}\frac{1}{1-x}dy dx = \int_{0}^{0.5}1 \ dx = \frac{1}{2}$$

$$P(E) = \frac{\frac{2log2-1}{2}}{\frac{1}{2}} = 2log2-1$$

Similarly $P(E)$ fomr case 2 will be $2log2-1$, the final probability will be $P(case1)P(E)+P(case2)P(E)$

Since the first break point was chosen at random from the interval $[0,1]$

$P(case1)=P_Z(0 \leq z \leq 0.5) = 0.5$

By similar arguments, $P(case2)=0.5$

Therefore, the final probability is $P(E) = 0.5*(2log2-1)+0.5*(2log2-1) = 2log2-1$

```
def plot_bar(d, ax):
    ax.bar(x=list(d.keys()), height=list(d.values()), linewidth=1, edgecolor='black')
    ax.yaxis.set_major_formatter(PercentFormatter(1))
    ax.set_yticks(np.arange(0,1.1,0.1))
    ax.set_ylim([0,1])
    ax.grid()

def get_dict(true, total):
    return {'True': true/total, 'False':(total-true)/total}
```

```
n = 1000
B = np.random.random(n)
```

```
fig, axes = plt.subplots(2,2, figsize=(8,8))
names = ['1/3<=B<=2/3', '|B-1/2|<=1/4', 'B<1/4 or 1-B<1/4', '3B^2 < B']
conditions = [get_dict(((1/3<=B) & (B <2/3)).astype(int).sum(), n)]
conditions.append(get_dict((np.abs(B-0.5)<1/4).astype(int).sum(), n))
conditions.append(get_dict(((B<1/4) | (1-B<1/4)).astype(int).sum(), n))
conditions.append(get_dict((3*B**2 < B).astype(int).sum(), n))
for idx, cond in enumerate(conditions):
    ax = axes[idx//2, idx%2]
    plot_bar(cond, ax)
    ax.set_title(names[idx])

plt.show()
```

```
def forms_triangle(p):
    l1 = p.min()
    l2 = 1-p.max()
    l3 = abs(p[0]-p[1])
    l = [l1, l2, l3]

    return True if all([l[i%3]+l[(i+1)%3]>= l[(i+2)%3] for i in range(3)]) else False
```

```
n = 10000
points = np.random.random((n,2))
true_cases = np.array([forms_triangle(p) for p in points]).astype(int).sum()
neg_cases = n - true_cases

plt.title('Probability of forming a triangle')
plt.bar(['True', 'False'], [true_cases/n, neg_cases/n], linewidth=1, edgecolor='black')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.yticks(np.arange(0,1.1,0.1))
plt.ylim([0,1])
plt.grid()
plt.show()
```

The estimated probability is 25%
